### Shifting Perspectives: *How The Pandemic Shaped Descriptions of Gratitude & the Year of 2020*

------------------

#### Kelsey Warren
May 7, 2021

Over the last year, the COVID-19 pandemic has created an environment where the things that we hold most precious are under threat. In the United States, we first heard about the virus and it felt like a distant threat, literally continents away. Then, as the virus got closer, our sense of security -- a privilege in our day and age where we are relatively sheltered from the health crises that happen on other continents -- was under threat. As the virus made landfall in the US, we were worried and a bit confused, carrying on our lives tentatively. As the days went on, and we were told that the virus was spreading rapidly, and our schools and workplaces were shut down. It was then that our places of community and connection were taken from us. We were told not to see friends and family. We were buckled down with our households, learning which masks were most safe and wrestling with how much we needed to wipe off the groceries. We were scared and hysterical: we hoarded toilet paper and beans, unsure how long this would go on, and feeling like we needed to make our residences into self-contained worlds.  As the weeks went on, we realized that it was not going to be a short-term obstacle to continuing our old lives. Soon, events months out were cancelled and we had to rearrange our lives around “WFH” and Zoom as our best friend. Outside our isolation at home, the devastation in the world was staggering: the case and death counts were rising dramatically every day in the United States, and around the world. We lost loved ones whom we couldn’t visit, and there was no immediate solution as we were told that the best we could do was social distance to keep ourselves and those around us safe before the vaccine arrived.

Over the last year, we've learned new ways of living and coping. The very things that we hold most dear -- our loved ones and health -- were and continue to be directly under threat. Given the unique challenges of the COVID-19 pandemic never before faced for most, there is reason to wonder if this monumental paradigm shift of social isolation and fear of sickness will change what people are most grateful for in their lives.

Given the unique dynamics around the high level of contagion of COVID-19 and the social isolation due to recovery efforts, there is reason to believe that people might value their loved ones going forward as it has become scarce under the conditions of the pandemic. Additionally, as our health has been under threat, there is reason to wonder whether people will be more grateful for their health going forward because it has been so central to how we’ve sought to avoid contagion in the last year.

### How can we measure expressions of gratitude around *loves ones* and *health*?

While we might be able to look at studies done on gratitude for the purposes of positive psychology, these sources would not provide us with the expressions of gratitude we need from recent months during the pandemic. Instead, we can look to social media to provide a snapshot of what people are grateful for. Namely, Twitter provides an interesting context to analyze text because it provides a way to look at gratitude from the level of individuals, posting out of their own autonomy, and thus (hopefully) reflective of their current state of mind on what they are grateful for. 

We want to make sure that we are looking at expressions of gratitude that are summary reflections of a year in review. This means that we want to look at times in the year when people are reflective. One such time would be during the holiday of Thanksgiving. In the United States, Thanksgiving Day comes along with the expectation that individuals take stock of what they are grateful for, and share that with those around them. Therefore, by comparing tweets on Thanksgiving 2019 and Thanksgiving 2020, we can evaluate whether people express gratitude differ after they’ve weathered 8 months of the pandemic. 

We will compare tweets on Thanksgiving Day for both years: November 28, 2019 and November 26, 2020. We want look at tweets in which the composer of the tweet is seeking to express gratitude, so we’ve selected tweets with words that trigger an expression of gratitude to some extent, such as *thanks*, *grateful*, *thankful*, *give thanks*, *appreciate*, etc. We have 15,000 tweets for each Thanksgiving day.


We are interested in exploring the question: **Are people more grateful for loved ones and health on Thanksgiving 2020 compared to Thanksgiving 2019?** My initial hypothesis would be that people are more grateful for both loved ones and health during the pandemic. 

### What do we find?

First, we can explore whether the topic of loved ones appears more in 2020 than in 2019. 

We find that *family* appears 1,710 times in the tweets from 2019, compared to 1,413 times in the tweets from 2020. The word *friends* appears 984 times in tweets on Thanksgiving 2019 compared to 767 times on Thanksgiving 2020. These frequencies appear to convey that loved ones show up more in 2019, but we need to look at these frequencies relative to the size of the number of words we have in tweets for each year. 

We can find the words are distinctive to the 2019 tweets compared to those in the 2020 tweets. We find that *family*, *friends*, and *#family* are more distinctive to the 2019 tweets when compared to the 2020 tweets. Therefore, we know that loved ones are more distinctive to 2019, even if they do not seem *that much* more frequent. 

We can look to see which words are distinctive to the context around the words *family* and *friends* to see how these words are being used in context to convey gratitude. In 2019, we find that these words are surrounded by positive descriptions such as *loving*,*wonderful*, and *amazing* more distinctively; and, in the 2020 tweets, the context that these words exist in is more qualifying of negative experiences with words like *through*, *but*, *sorry*, *tough*, and *crazy*. Perhaps, this reflects the idea that we have more exciting associations with friends and family in 2019 when we had experiences in person with them; in 2020, our friends and family were more distant so people express gratitude for them in the context of the acknowledgment of the challenges of the year. 

Next, we can explore the topic of health. We can look at invocations of the topic of health through looking at the frequency and contexts of the words *health* and *healthy.* We see that *health* appears 141 times in the tweets on Thanksgiving 2019 compared to 203 times on Thanksgiving 2020. We can also see that the word *healthy* appears 52 times in the tweets on Thanksgiving 2019 compared to 109 times on Thanksgiving 2020. We find that these two words are more distinctive to the 2020 tweets compared to 2019. However, what is striking is that this topic is not being invoked nearly as much as loved ones are. This suggests that people are much more likely to talk about friends and family on Thanksgiving.

Similar to discussions of loved ones, we see that *health* exists in different contexts in 2019 and 2020. In 2019, *health* exists in the context of words that are distinctive to these tweets, like *best*, *great*, *abundance*, *life*; however, in 2020, it exists in the context of words related to the pandemic, such as *safe*, *healthcare*, *care*, and modifiers that convey a negative context, such as *despite*, *but*, *despite*.  This suggests that expressions of gratitude in 2020 are more often couched with phrases that qualify gratitude in spite of something; for example, “I am grateful for my family, despite being separated due to the pandemic.” In this way, we can see how gratitude actually functions differently in the different years. In 2019, expressions of gratitude appear as a positive declaration itself; this is compared to 2020 when they exist alongside the descriptions of a challenge, where the writers feel the need to make it known that something is happening in a less-than-ideal context.

We see how the context of mentions of health enhances the gratitude in 2019; whereas in 2020, it occurs in the context of descriptions of the pandemic and/or in spite of challenges, in essence taking away from the subject of the gratitude by giving breath to the clause after the *but* or *despite*.

What these analyses tell us is that the tweets of these two years differ in how gratitude is framed. We can confirm these findings around positive versus negative descriptions in tweets between 2019 and 2020 by looking at the adjectives used in the two sets of tweets. The unique adjectives of 2019 are positive words, such as *wonderful*, *amazing*, *grateful*, and *thankful*. Distinctive adjectives to 2020 are those related to COVID (*safe*, *healthy*, *normal*) and those relating to challenges faced (*different*, *difficult*, *sorry*, *weird*, *sad*, *tough*, *crazy*). 

To put all this together, for 2019, we have adjectives enhancing the gratitude, and for 2020 we have negative adjectives and conjunctions communicating *in spite* of something negative. Perhaps, our textual analysis becomes reflective of the experience of 2020. In 2019, people are grateful for things because they are positive things, and in 2020 they are grateful in spite of hardship.

### So what does this mean for our hypothesis?

Our hypothesis is proven false: we find evidence that the topics of loved ones are actually more distinctive to tweets on Thanksgiving 2019 and the topic of health to tweets on Thanksgiving 2020. Perhaps, this reflects an availability bias that these two topics are at different levels of immediacy to composers of tweets in the different years. In 2019, we were more surrounded by family and friends, so we are more eager to write about them and our experiences with them. In 2020, we are surrounded by personal experiences and a media environment all focused on the pandemic, so we refer to health more often. 

Through looking at the context of both of these topics, we see how the tweets from Thanksgiving 2019 situate these topics in descriptions that are positive and bring focus to the topic itself. In 2020, we see how the pandemic plays a central role in what people are commenting on; the fact that these expressions of gratitude are occurring in the context of mentioning something else after a conjunction suggests these topics of gratitude are almost secondary to the topic of the year and the pandemic.

# 2019 vs. 2020.

It has become clear through our analysis that the distinctive textual contexts we’ve found in tweets in the two years are reflective of the experience of the years themselves. For many, 2019 was a lot more optimistic than 2020. Also, in 2020, there is a need to recognize our blessings but also pay homage to the tragedy of the year. 

What's more, in popular culture, the idea of “2020” has become somewhat of a trope, synonymous with the challenges of the year. We can look at some popular memes from this year to get a sense of this phenomenon. 

We can look at whether people actually talk about these years differently as they write about gratitude on Thanksgiving.

### How do these tweets reflect the recognition of the uniqueness of 2020 as compared to 2019?

We can look directly at mentions of *2019* versus *2020*. Strikingly, we find that *2020* shows up over twice the amount of time in tweets on Thanksgiving 2020 than *2019* shows up on Thanksgiving 2020. This suggests that people really are attached to this idea of the calendar year 2020 as a way to categorize the trauma of the year.

We see the same general trend show up with mentions of the years 2019 vs. 2020 as it did in the context of the topics we explored above. The context of the word *2019* is more focused on optimistic depictions of words that convey gratitude, Thanksgiving, and big picture ideas like *hope* and *love*. The most distinctive word to the context of the use of *2020* is *safe* which reflects how people want to comment on safety as a catch-all for desiring stability and solace amidst the chaos of the year.

We can look further at this idea by looking at the words around the word *year* to compare how people reflect on the two years in question. The words that surround *year* in 2019 are more positive -- *wonderful*, *amazing* -- and indicative of gratitude and Thanksgiving themes (*grateful*, *thankful*, *thanksgiving*). In 2020, the distinctive words around *year* were about enduring the year (*during*, *through*, *been*, *stay*) and descriptions of its relatedness as comparatively worse than other years (*different*, *weird*, *crazy*, *tough*).

The crucial takeaway here is that people are invoking both more vague and optimistic verbiage about the year in 2019 compared to the seemingly challenged and practical verbiage of the experiences of 2020. For most people, the years of 2019 and 2020 were as different as they come. It follows then, that descriptions of the years themselves appear in different contexts as people have different associations with the year when they have had different experiences during them.

### Conclusion

We set out to explore whether the paradigm shift that the COVID-19 pandemic has caused in our world also shifted how we express gratitude on Twitter on Thanksgiving. We have found that composers of the tweets sampled are conscious of the different world we have come to inhabit. 

In 2019, people were more likely to comment on friends and family. In 2020, people were more likely to talk about health. We see how people see the world in less glowing terms and describe things more negatively in 2020 when compared to 2019. We also see this tension between expressing gratitude (which has positive goals) and the need to recognize that this gratitude is occurring in the context of challenges. Significantly, we see how people view the year in which they are writing the tweet in vastly different ways, reflecting the challenges experienced in the real world. In this way, we can see the power of language to reflect peoples' dynamic and complex experiences of reality. 

Looking forward, as we still wrestle with the challenges of this virus, we can only hope that we can maintain much to be grateful for. And, most importantly, that we continue to take to Twitter to share the good cheer.
