## How Has the COVID-19 Pandemic Changed What People Express Gratitude For?
### Using Twitter Data To Compare Expressions of Gratitude Between Thanksgiving 2019 & 2020.

#### COMM313 Final Project - Spring 2021 - Kelsey Warren  


## Research Question:   
When tweeting about what they are grateful for on Thanksgiving, do people tweet more about loved ones and health on Thanksgiving 2020 (during the pandemic) compared to Thanksgiving 2019?

## Hypothesis:   
Tweets on Thankgiving in 2020 will refer more to: 1. health; 2. loved ones (family, friends, partners) than those in 2019, before the COVID pandemic.  

## The data: 

### Where from? How did I get it?

Used Twitter API to get 1000 tweets for 8 queries on Thanksgiving day for November 28, 2019 and November 26, 2020, so I have 8000 tweets per year = 16000 total.

    - I am thankful for 
    - I am grateful for
    - This year I am thankful for 
    - This year
    - Thankful 
    - Grateful 
    - Gratitude 
    - Give thanks

## How did I get it/ is it organized? 

It is organized in two folder by year, and then 8 separate files by the query.
